## Segmentation_Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os
```

```python
# Ensure images folder exists
os.makedirs("images", exist_ok=True)

# --- Data setup ---
data = {
    "Metric": ["Day 1 Retention", "Day 7 Retention", "Day 30 Retention"],
    "Your Game": [34.9, 17.2, 9.4],
    "Industry Benchmark (Low)": [30, 10, 3],
    "Industry Benchmark (High)": [40, 20, 8]
}

bench_df = pd.DataFrame(data)

# --- Plot ---
plt.figure(figsize=(8, 5))
sns.barplot(x="Metric", y="Your Game", data=bench_df, color="steelblue", label="Your Game")

# Overlay benchmark ranges
plt.errorbar(
    x=range(len(bench_df)),
    y=(bench_df["Industry Benchmark (Low)"] + bench_df["Industry Benchmark (High)"]) / 2,
    yerr=(bench_df["Industry Benchmark (High)"] - bench_df["Industry Benchmark (Low)"]) / 2,
    fmt='o',
    color='orange',
    capsize=8,
    label="Industry Range"
)

# Style
plt.title("Retention Comparison – The Game vs. Industry Benchmarks", fontsize=14, weight='bold')
plt.ylabel("Retention (%)")
plt.xlabel("")
plt.ylim(0, 45)
plt.legend()
plt.tight_layout()

# Save chart
bench_path = "images/retention_vs_benchmark.png"
plt.savefig(bench_path, dpi=300)
plt.show()

print(f"Retention benchmark chart saved to: {bench_path}")
```
